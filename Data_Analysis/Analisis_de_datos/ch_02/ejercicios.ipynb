```python
#Ejercicio 1
df1=df.loc[(df.parsed_place=="Japan") & (df.magType=="mb"),["mag"]]#filtramos
df1.info()#comprobamos que no haya nulos
```

```python
df1.quantile(0.95) # calculamos el percentil 95
```

```python
#Ejercicio 2
df2=df.loc[(df.parsed_place=="Indonesia"),["tsunami"]]
df2.info()
```

```python
#Porcentaje de terremotos acompañados de tsunami en Indonesia
df2.sum()/df2.count()*100
```

```python
#Ejercicio 3
df.loc[(df.parsed_place=="Nevada")].describe()
```

```python
#Ejercicio 4

#creamos lista con países/estados pertenecientes al Cinturón de Fuego
c_fuego=["Alaska","Antarctic","Bolivia","California","Canada","Chile","Costa Rica","Ecuador","Fiji","Guatemala","Indonesia","Japan","Kermadec Islands","Mexico","New Zealand","Peru","Philippines","Russia","Taiwan","Tonga","Washington"]

df["cinturon_fuego"]=df.parsed_place.apply(lambda x: 1 if x in c_fuego else 0) #creamos columna y añadimos 1 si el lugar está en la lista y 0 en caso contrario
df
```

Calcula el número de terremotos en los lugares del Cinturón de Fuego y el número fuera de ellos.

```python
#Ejercicio 5

#Numero de terremotos en el Cinturón de Fuego
df.cinturon_fuego.sum()
```

```python
#Numero de terremotos fuera del Cinturón de Fuego
df.shape[0]-df.cinturon_fuego.sum()
```

```python
#Ejercicio 6
df.loc[(df.cinturon_fuego==1),["tsunami"]].sum() #numero de tsunamis en el Cinturón de Fuego
```
